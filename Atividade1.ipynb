```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Importando o dataFrame
df = pd.read_csv('Atividade1.csv', engine='python', sep=';', decimal=',')
display(df)
```

```python
velocidade = ['0-19',	'20-29', '30-39',	'40-49', '50-59',	'60-69', '70-79',	'80-89', '90-99', '100-109', '110-119',	'120-129', '130-139',	'140-199']
total = [1626, 4455, 4794, 5129, 49489, 423000, 523715, 8049, 299, 115, 71, 21, 4, 1]
media_velocidade = [9.5, 24.5, 34.5, 44.5, 54.5, 64.5, 74.5, 84.5, 94.5, 104.5, 114.5, 124.5, 134.5, 169.5]
```

```python
# Criando novo dataFrame a partir dos dados do data frame anterior
novoFrequencia_df = pd.DataFrame({'Velocidade': velocidade,
                        'Total': total})
```

```python
novoFrequencia_df
```

```python
novoFrequencia_df.info()
```

```python
# Criando um novo data Frame agora contendo a média das velocidades na coluna Velocidade
novoMedia_df = pd.DataFrame({'Media Velocidade': media_velocidade,
                             'Total': total})
```

```python
novoMedia_df
```

```python
#Para calcular a média fazer o produto da média de velocidade pela frequência, dividido pelo total das frequências
mediaVelocidades = (novoMedia_df['Velocidade'] * novoMedia_df['Total']).sum() / novoMedia_df['Total'].sum()
mediaVelocidades
```

```python
# Para calcular a variância utilizar o método var() para retornar o resultado
varianciaVelocidades = novoMedia_df['Velocidade'].var()
varianciaVelocidades
```

```python
#O desvio-padrão para distribuição de frequências precisa levar em conta o número de vezes que cada medida foi feita (frequência)
desvioPadraoVel = novoMedia_df['Velocidade'].std()
desvioPadraoVel
```

```python
to_plot = plt.bar(novoFrequencia_df['Velocidade'],novoFrequencia_df['Total'])
plt.xticks(rotation=90)
plt.show()
```
